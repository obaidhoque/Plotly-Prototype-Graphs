```python
data1 = pd.read_csv("/Users/OBAID/Desktop/stat.csv")
data1.rename(columns = { 'RawData.Day' : 'col1',
                       'MeanMinus3Std' : 'col2'}, inplace = True)
data1.head(2)
```

```python
data1points = go.Scattergl(x = data1.col2,
                          y = np.log(data1.Mean),
                          mode = 'markers',
                          marker = dict(
                          opacity=0.75,
                          line=dict(
                          color='red'
                          )
                          ),
                          name = 'Data points'
                          )

m,b = np.polyfit(data1.col2, np.log(data1.Mean), 1)
bestfit_y = (data1.col2 * m + b)

lineOfBestFit=go.Scattergl(
    x=data1.col2,
    y=bestfit_y,
    name='Line of best fit',
    line=dict(
        color='blue',
    )
)

data=[data1points, lineOfBestFit]

layout = go.Layout(
    title='Mean vs Mean minus 3rd Deviation',
    xaxis=dict(
        title='Mean'
    ),
    yaxis=dict(
        title='Mean - 3rd Deviation'
    ),
    hovermode='closest',
    #showlegend=True
)


figure = go.Figure(data=data, layout=layout)

figure.show()
```
